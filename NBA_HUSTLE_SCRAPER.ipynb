```python
## NBA Shot distance Stats scrape 
from selenium import webdriver
from bs4 import BeautifulSoup as soup 
import pandas as pd
from selenium.webdriver.support.ui import Select
import re
```

```python
### connecting to web driver
driver = webdriver.Chrome()
##MISC stats
driver.get('https://stats.nba.com/players/hustle/?Season=2019-20&SeasonType=Regular%20Season')

 
##select season drop down 
Season_element= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select'))
##uncomment the year you want
Season_2019_2020= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(1)").click()
##Season_2018_2019=driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(2)").click()
#Season_2017_2018=driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(3)").click()
##Season_2016_2017=driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(4)").click()
##Season_2015_2016=driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(5)").click()
##Season_2014_2015=driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > div.row.row5.collapse.stats-filters-top > div:nth-child(1) > div > div > label > select > option:nth-child(6)").click()

```

```python
### find drop down and select
ddelement= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select'))

##find all and select
All_Element= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select > option:nth-child(1)").click()
```

```python
table=driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr')
links = []
import collections
for telement in table:
    table_dict = collections.OrderedDict()
    link = telement.find_element_by_xpath('.//td[1]/a').get_attribute('href') + 'misc/'
    Player = telement.find_element_by_xpath('.//td[1]/a').text
    Team = telement.find_element_by_xpath('.//td[2]/a').text
    Age = telement.find_element_by_xpath('.//td[3]').text
    GP = telement.find_element_by_xpath('.//td[4]').text
    Min = telement.find_element_by_xpath('.//td[5]').text
    Screen_Assists= telement.find_element_by_xpath('.//td[6]').text
    Screen_Assists_Pts= telement.find_element_by_xpath('.//td[7]').text
    Deflections= telement.find_element_by_xpath('.//td[8]').text
    Off_loose_balls_recovered= telement.find_element_by_xpath('.//td[9]').text
    Def_loose_balls_recovered= telement.find_element_by_xpath('.//td[10]').text
    loose_balls_recovered= telement.find_element_by_xpath('.//td[11]').text
    Off_loose_balls_recovered_pct=telement.find_element_by_xpath('.//td[12]').text
    Def_loose_balls_recovered_pct=telement.find_element_by_xpath('.//td[13]').text
    Charges_drawn=telement.find_element_by_xpath('.//td[14]').text
    Contested_2p_shots=telement.find_element_by_xpath('.//td[15]').text
    Contested_3p_shots=telement.find_element_by_xpath('.//td[16]').text
    Contested_shots=telement.find_element_by_xpath('.//td[17]').text
    links.append([link, Player,Team,Age,GP,Min, Screen_Assists,Screen_Assists_Pts,Deflections,
                  Off_loose_balls_recovered,Def_loose_balls_recovered,loose_balls_recovered,Off_loose_balls_recovered_pct,
                  Def_loose_balls_recovered_pct,Charges_drawn, Contested_2p_shots, Contested_3p_shots,Contested_shots])
```

```python
###converting the scraped list to a dataframe
Hustle_Stats=pd.DataFrame(links)
```

```python
#### naming the columns
Hustle_Stats.columns= ["link","Player","Team","Age","GP","Min","Screen_Assists","Screen_Assists_Pts","Deflections",
                  "Off_loose_balls_recovered","Def_loose_balls_recovered","loose_balls_recovered","Off_loose_balls_recovered_pct",
                  "Def_loose_balls_recovered_pct","Charges_drawn","Contested_2p_shots","Contested_3p_shots","Contested_shots"]
```

```python
Hustle_Stats.to_csv("Hustle_Stats_17.csv")
```
