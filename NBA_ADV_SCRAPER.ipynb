```python
### find drop down and select
ddelement= Select(driver.find_element_by_css_selector('body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select'))

##find all and select
All_Element= driver.find_element_by_css_selector("body > main > div.stats-container__inner > div > div.row > div > div > nba-stat-table > div:nth-child(1) > div > div > select > option:nth-child(1)").click()
```

```python
##plyr = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[1]/div[1]/table/tbody/tr')
table=driver.find_elements_by_xpath('/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[2]/div[1]/table/tbody/tr')
links = []
import collections
for telement in table:
    table_dict = collections.OrderedDict()
    link = telement.find_element_by_xpath('.//td[2]/a').get_attribute('href') + 'advanced/'
    Player = telement.find_element_by_xpath('.//td[2]/a').text
    Team = telement.find_element_by_xpath('.//td[3]/a').text
    Age = telement.find_element_by_xpath('.//td[4]').text
    GP = telement.find_element_by_xpath('.//td[5]').text
    W = telement.find_element_by_xpath('.//td[6]').text
    L = telement.find_element_by_xpath('.//td[7]').text
    Min = telement.find_element_by_xpath('.//td[8]').text
    Offrtg = telement.find_element_by_xpath('.//td[9]').text
    Defrtg = telement.find_element_by_xpath('.//td[10]').text
    Netrtg = telement.find_element_by_xpath('.//td[11]').text
    Ast_Pct = telement.find_element_by_xpath('.//td[12]').text
    Ast_TO=telement.find_element_by_xpath('.//td[13]').text
    Ast_Ratio=telement.find_element_by_xpath('.//td[14]').text
    OREB_Pct=telement.find_element_by_xpath('.//td[15]').text
    DREB_Pct=telement.find_element_by_xpath('.//td[16]').text
    REB_Pct=telement.find_element_by_xpath('.//td[17]').text
    To_Ratio=telement.find_element_by_xpath('.//td[18]').text
    EFG_Pct=telement.find_element_by_xpath('.//td[19]').text
    TS_Pct=telement.find_element_by_xpath('.//td[20]').text
    USG_Pct=telement.find_element_by_xpath('.//td[21]').text
    Pace=telement.find_element_by_xpath('.//td[22]').text
    PIE=telement.find_element_by_xpath('.//td[23]').text
    links.append([link, Player,Team,Age,GP,W,L,Min,Offrtg,Defrtg,Netrtg,
                  Ast_Pct,Ast_TO,Ast_Ratio,OREB_Pct,DREB_Pct,REB_Pct,To_Ratio,EFG_Pct,TS_Pct,USG_Pct,Pace,PIE])
```

```python
###converting the scraped list to a dataframe
ADV_Stats=pd.DataFrame(links)
```

```python
#### naming the columns
ADV_Stats.columns= ["link","Player","Team","Age","GP","W","L","Min","Offrtg","Defrtg","Netrtg",
"Ast_Pct","Ast_TO","Ast_Ratio","OREB_Pct","DREB_Pct","REB_Pct","To_Ratio","EFG_Pct","TS_Pct",
"USG_Pct","Pace","PIE"]
```
